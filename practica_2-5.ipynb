## ✅ Recordemos la matriz del ejercicio 1.c:

$$
A = \begin{bmatrix}
1 & 1 & 0 \\
0 & 1 & 0 \\
0 & 1 & 1 \\
\end{bmatrix}
$$

Queremos escribirla como $A = LU$, donde:

* $L$ es una **matriz triangular inferior** con **unos en la diagonal**.
* $U$ es una **matriz triangular superior**.

---

## ✏️ Paso a paso: Factorización LU **(sin pivoteo)**

Vamos a hacer lo mismo que en la eliminación gaussiana para triangular la matriz, **pero guardando los coeficientes usados** en las filas para formar la matriz $L$.

### Paso 1: Inicializamos $L$ y $U$

Partimos con:

$$
L = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}, \quad
U = A = \begin{bmatrix}
1 & 1 & 0 \\
0 & 1 & 0 \\
0 & 1 & 1 \\
\end{bmatrix}
$$

---

### Paso 2: Hacer ceros debajo del pivote en la columna 1

Ya está hecho: debajo del $1$ en (1,1), hay un $0$ en (2,1) y un $0$ en (3,1), así que no hay que hacer nada.
Por lo tanto, en la matriz $L$, los elementos $L_{21}$ y $L_{31}$ son **0**.

---

### Paso 3: Hacer ceros debajo del pivote en la columna 2

Queremos hacer 0 en (3,2).
Ya tenemos un 1 en (2,2), así que:

$$
\text{Multiplicador} = \frac{U_{32}}{U_{22}} = \frac{1}{1} = 1
$$

Entonces:

$$
F3 \leftarrow F3 - 1 \cdot F2
$$

Resultado en $U$:

$$
U =
\begin{bmatrix}
1 & 1 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}
$$

Y el valor del multiplicador lo ponemos en $L_{32} = 1$:

$$
L =
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 1 & 1 \\
\end{bmatrix}
$$

---

## ✅ Resultado final:

$$
L = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 1 & 1 \\
\end{bmatrix}, \quad
U = \begin{bmatrix}
1 & 1 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}
$$

Y se cumple que $A = LU$.

---

## ℹ️ ¿Cuándo se necesita **pivoteo**?

Se necesita **pivoteo** (intercambiar filas) cuando:

* El elemento en la **diagonal** es **cero** (por ejemplo, $A_{11} = 0$), o
* Es un número muy pequeño y puede generar errores numéricos al dividir (si estás usando este método en la computadora).

En esos casos, se intercambian filas para que el nuevo pivote no sea cero.
