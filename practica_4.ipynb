## Práctica 4. Modelos de regresión. 

Para ajustar una recta a los datos usando el **método de los mínimos cuadrados**, buscamos una recta de la forma:

$$
y = a x + b
$$

Dada la tabla:

| x | 0 | 1 | 2 | 3 |
| - | - | - | - | - |
| y | 1 | 2 | 3 | 6 |

### Paso 1: Calcular las sumas necesarias

Sean $n = 4$ datos.

$$
\sum x = 0 + 1 + 2 + 3 = 6 \\
\sum y = 1 + 2 + 3 + 6 = 12 \\
\sum x^2 = 0^2 + 1^2 + 2^2 + 3^2 = 14 \\
\sum xy = 0\cdot1 + 1\cdot2 + 2\cdot3 + 3\cdot6 = 26
$$

### Paso 2: Fórmulas de mínimos cuadrados

$$
a = \frac{n\sum xy - \sum x \sum y}{n\sum x^2 - (\sum x)^2} = \frac{4\cdot26 - 6\cdot12}{4\cdot14 - 6^2} = \frac{104 - 72}{56 - 36} = \frac{32}{20} = 1.6
$$

$$
b = \frac{\sum y - a \sum x}{n} = \frac{12 - 1.6\cdot6}{4} = \frac{12 - 9.6}{4} = \frac{2.4}{4} = 0.6
$$

### Resultado

La recta ajustada por mínimos cuadrados es:

$$
\boxed{y = 1.6x + 0.6}
$$


2 – ¿Cuál es el punto del plano dado por x + y - z = 0 más cercano al punto (2, 1, 0)? 

Queremos encontrar el **punto del plano $x + y - z = 0$** que está **más cercano al punto $P = (2, 1, 0)$**. Esto es un problema clásico de geometría analítica o cálculo vectorial, y se resuelve **proyectando el punto sobre el plano**.

---

### 🔧 Paso 1: Normal del plano

El plano $x + y - z = 0$ tiene como **vector normal**:

$$
\vec{n} = (1, 1, -1)
$$

---

### 🧭 Paso 2: Proyección del punto sobre el plano

La fórmula para proyectar un punto $P$ sobre un plano con normal $\vec{n}$ es:

$$
P_{\text{proy}} = P - \frac{(P \cdot \vec{n} - d)}{||\vec{n}||^2} \cdot \vec{n}
$$

Donde:

* $P = (2, 1, 0)$
* $\vec{n} = (1, 1, -1)$
* El plano es $x + y - z = 0 \Rightarrow d = 0$

1. Producto punto $P \cdot \vec{n} = 2 + 1 - 0 = 3$
2. Norma al cuadrado de $\vec{n}$:
   $||\vec{n}||^2 = 1^2 + 1^2 + (-1)^2 = 3$

$$
P_{\text{proy}} = (2, 1, 0) - \frac{3 - 0}{3} \cdot (1, 1, -1) = (2, 1, 0) - 1 \cdot (1, 1, -1)
$$

$$
P_{\text{proy}} = (2 - 1, 1 - 1, 0 - (-1)) = (1, 0, 1)
$$

---

### ✅ Resultado final

El punto del plano más cercano a $(2, 1, 0)$ es:

$$
\boxed{(1, 0, 1)}
$$


3 – Comprobar que en regresión lineal con cuadrados mínimos, cuando los puntos son 2 se
obtiene precisamente la recta que pasa por ambos puntos (¡para esto agregar una condición
indispensable sobre ambos puntos!). Más generalmente, se puede probar que si los puntos
están alineados (haya o no haya repeticiones), se obtendrá la recta que los contiene. ¿Por
qué? (Explicar, sin hacer cuentas.) 

### ✅ **Parte 1: Con solo 2 puntos, la recta por mínimos cuadrados pasa por ambos**

Supongamos que tenemos dos puntos distintos:

$$
(x_1, y_1), \quad (x_2, y_2) \quad \text{con } x_1 \ne x_2
$$

Esta condición $x_1 \ne x_2$ es indispensable. Si no se cumple, no se puede ajustar una recta (sería una línea vertical o los puntos se superponen).

**¿Por qué la recta por mínimos cuadrados pasa por ellos?**
Porque con solo 2 puntos distintos, **hay una única recta que los contiene**. Y al aplicar el método de los mínimos cuadrados, estamos buscando **la recta que minimiza la suma de los errores al cuadrado** entre las predicciones y los valores reales. Pero como solo hay dos datos, la única forma de minimizar esa suma de errores es **que ambos errores sean cero**, es decir, que la recta **pase exactamente por ambos puntos**.

---

### ✅ **Parte 2: Si los puntos están alineados, la recta por mínimos cuadrados los contiene**

**¿Qué significa que los puntos estén alineados?**
Que todos están sobre la misma recta, es decir, hay $y = ax + b$ tal que cada punto $(x_i, y_i)$ la cumple.

**¿Por qué la regresión da esa misma recta?**
Porque el método de mínimos cuadrados busca **la recta que mejor se ajusta a los puntos**, minimizando la suma:

$$
\sum (y_i - (ax_i + b))^2
$$

Estas 2 formas son iguales porque despues lo elebas al cuadrado 

$$
\sum (ax_i + b - y_i)^2
$$

Si los puntos ya están todos **exactamente sobre una recta**, entonces esa suma es **cero** para esa recta.
Y como no hay forma de obtener un valor menor que cero, el método **encontrará esa misma recta** como solución.

---

### ✳️ Conclusión

* Con **2 puntos distintos**, hay una única recta que pasa por ellos, y es la que da mínimos cuadrados.
* Si los puntos están **alineados**, la recta que los contiene es la única con **error total cero**, y por eso el método la encuentra.


Ejemplo simple con **3 puntos alineados**, y se vera que la **regresión lineal por mínimos cuadrados encuentra la recta que pasa exactamente por los 3**.

### 🎯 Ejemplo

Tomemos los siguientes puntos:

$$
(1, 2), \quad (2, 4), \quad (3, 6)
$$

Claramente, todos están sobre la recta $y = 2x$.

### 🧮 Apliquemos mínimos cuadrados

Queremos encontrar la recta:

$$
y = ax + b
$$

Usamos las fórmulas:

$$
a = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} \quad ; \quad b = \frac{\sum y - a \sum x}{n}
$$

---

$$
\sum x = 6, \quad \sum y = 12, \quad \sum x^2 = 14, \quad \sum xy = 28, \quad n = 3
$$

Calculamos:

$$
a = \frac{3 \cdot 28 - 6 \cdot 12}{3 \cdot 14 - 6^2} = \frac{84 - 72}{42 - 36} = \frac{12}{6} = 2
$$

$$
b = \frac{12 - 2 \cdot 6}{3} = \frac{12 - 12}{3} = 0
$$

---

### ✅ Resultado

La recta ajustada es:

$$
\boxed{y = 2x}
$$

Y efectivamente, **pasa por los 3 puntos exactos**. Por lo tanto, **la suma de los errores es cero**. Si haces la suma de los 3 puntos dados, vez que es 0 porque cada uno de esos calculos es 0

$$
\boxed{\sum (ax_i + b - y_i)^2}
$$

4 – Para regresión lineal con cuadrados mínimos con el objetivo de hallar la recta de
pendiente a y ordenada al origen b, mostrar una condición tal que si valiera el a o el b
quedarán indefinidos

Buena pregunta. En regresión lineal por mínimos cuadrados, **hay una situación en la que los coeficientes $a$ (pendiente) y $b$ (ordenada al origen)** pueden quedar **indefinidos o indeterminados**.

---

## 🧠 Fórmulas recordadas:

La regresión lineal encuentra la recta:

$$
y = ax + b
$$

con:

$$
a = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}
\quad\quad
b = \frac{\sum y - a \sum x}{n}
$$

---

## ❗️Condición que hace que $a$ y $b$ queden indefinidos

### 👉 Si **todos los valores de $x$** son iguales (es decir, $x_1 = x_2 = \cdots = x_n$):

Entonces:

* $\sum x^2 = x^2 + x^2 + \cdots = nx^2$
* $(\sum x)^2 = (nx)^2 = n^2x^2$

Y el **denominador de la fórmula de $a$**:

$$
n \sum x^2 - (\sum x)^2 = n(nx^2) - (nx)^2 = n^2x^2 - n^2x^2 = 0
$$

$$
\Rightarrow a \text{ está indefinido (división por 0)}.
$$

Como $a$ es indefinido, entonces $b$ también **queda indefinido** porque depende de $a$.

---

### 🧱 Intuición geométrica:

Si todos los valores de $x$ son iguales, entonces todos los puntos están **sobre una línea vertical**.

Ejemplo:

$$
(3, 1), (3, 2), (3, 7)
$$

En este caso no tiene sentido ajustar una **recta de la forma $y = ax + b$**, porque ninguna recta de esa forma **puede ser vertical**.
Este es el motivo por el cual el ajuste **falla o queda indefinido**.

---

### ✅ Conclusión:

> **La condición para que $a$ y $b$ queden indefinidos es que todos los valores de $x$ sean iguales.**




## ✅ 1. ¿La recta no puede ser vertical porque la pendiente sería infinita?

**Sí, exactamente.**
Una **recta vertical** tiene **pendiente indefinida** (o "infinita", en cierto sentido informal), porque para calcular la pendiente usamos:

$$
a = \frac{\Delta y}{\Delta x}
$$

Si los puntos tienen el mismo $x$, entonces $\Delta x = 0$, y:

$$
a = \frac{\text{algo}}{0} \Rightarrow \text{indefinido}
$$

Por eso, en el contexto de la regresión clásica que busca $y = ax + b$, **no se pueden representar rectas verticales**.

---

## ✅ 2. ¿Otra condición para que $a$ y $b$ estén indefinidos es que los puntos sean todos iguales? O directamente que solo se tenga un punto

**¡Correcto también!**

Si todos los puntos son **idénticos**, por ejemplo:

$$
(2, 5), (2, 5), (2, 5)
$$

Entonces:

* Todos los $x$ son iguales ⇒ ya sabés que el denominador del cálculo de $a$ se hace cero.
* Y además, **no hay variación en los datos**, por lo tanto **no hay forma de ajustar una recta única**, ya que hay **infinitas rectas** que pasan por un solo punto (todas las que pasan por $(2, 5)$).

### Entonces:

> **Sí: si todos los puntos son iguales**, también se hace **indefinida la regresión** porque no hay variación para explicar, y además el cálculo de $a$ implica dividir por cero.

---

## 🔍 Resumen:

| Situación                     | ¿Qué pasa?                                   |
| ----------------------------- | -------------------------------------------- |
| Todos los $x$ son iguales     | La pendiente $a$ es indefinida               |
| Todos los puntos son iguales  | $a$ y $b$ son indefinidos (no hay variación) |
| Recta vertical (tipo $x = c$) | No puede representarse como $y = ax + b$     |

Para aplicar **regresión lineal por mínimos cuadrados** y encontrar los coeficientes de la recta $y = ax + b$, necesitamos resolver un sistema de ecuaciones lineales, y una forma de hacerlo es con **la regla de Cramer**, la cual **requiere que el determinante de la matriz de coeficientes sea distinto de cero**.

---

### 📌 ¿Cuándo quedan indefinidos $a$ o $b$?

Una situación en la que **los coeficientes $a$ y $b$ quedan indefinidos** es cuando:

#### 🔴 Todos los valores de $x$ son iguales

(Sólo un punto o varios con la misma coordenada $x$).

---

### 🧮 Justificación con sistema lineal y determinante

En regresión lineal, el sistema que se resuelve con Cramer se basa en:

$$
\begin{cases}
a \sum x_i^2 + b \sum x_i = \sum x_i y_i \\
a \sum x_i + b n = \sum y_i
\end{cases}
$$

La matriz de coeficientes es:

$$
A = \begin{bmatrix}
\sum x_i^2 & \sum x_i \\
\sum x_i   & n
\end{bmatrix}
$$

El **determinante** es:

$$
\text{det}(A) = \left( \sum x_i^2 \cdot n \right) - \left( \sum x_i \right)^2
= n \cdot \sum x_i^2 - \left( \sum x_i \right)^2
$$

Esto es exactamente:

$$
n \cdot \text{Var}(x)
$$

> 🔹 Por lo tanto, **si la varianza de los $x_i$ es cero**, el determinante es cero y no se puede usar Cramer, ni resolver el sistema: **$a$ y $b$ quedan indefinidos.**

---

### ✅ Condición para que $a$ o $b$ queden indefinidos

**Condición:**

> Si todos los valores de $x_i$ son iguales (o solo hay un punto), entonces
>
> $$
> \text{Var}(x) = 0 \Rightarrow \text{det}(A) = 0
> $$
>
> y no se puede aplicar la regresión lineal con este método.

---

### 📌 Ejemplo:

Supongamos solo un punto: $(2, 5)$

Entonces:

* $\sum x_i = 2$
* $\sum x_i^2 = 4$
* $\sum y_i = 5$
* $\sum x_i y_i = 10$
* $n = 1$

$$
\text{det}(A) = 1 \cdot 4 - (2)^2 = 4 - 4 = 0
$$

6 – Para regresión lineal, comprobar que si se aplica una escala de h a todos los xi, i.e. se
cambia el valor xi por h.xi para 0 ≤ i ≤ n, el resultado de la regresión será el cambio del
coeficiente a por a / h. Interpretar este hecho en forma práctica.


Dado un conjunto de puntos $(x_i, y_i)$, queremos ver qué ocurre si **multiplicamos todos los $x_i$ por un factor $h$**:

$$
x_i \rightarrow h \cdot x_i
$$

¿Qué pasa con la recta de regresión $y = ax + b$?

---

#### Fórmula del coeficiente $a$

La pendiente $a$ en regresión lineal por mínimos cuadrados es:

$$
a = \frac{n \sum x_i y_i - \sum x_i \sum y_i}{n \sum x_i^2 - (\sum x_i)^2}
$$

Si cambiamos cada $x_i$ por $h x_i$, analizamos cómo cambia la fórmula:

#### Sustituimos $x_i \rightarrow h x_i$

* $\sum x_i \rightarrow h \sum x_i$
* $\sum x_i^2 \rightarrow h^2 \sum x_i^2$
* $\sum x_i y_i \rightarrow h \sum x_i y_i$

Sustituimos en la fórmula:

$$
a' = \frac{n(h \sum x_i y_i) - (h \sum x_i)(\sum y_i)}{n(h^2 \sum x_i^2) - (h \sum x_i)^2}
= \frac{h(n \sum x_i y_i - \sum x_i \sum y_i)}{h^2(n \sum x_i^2 - (\sum x_i)^2)}
$$

Simplificamos:

$$
a' = \frac{1}{h} \cdot \frac{n \sum x_i y_i - \sum x_i \sum y_i}{n \sum x_i^2 - (\sum x_i)^2} = \frac{a}{h}
$$

#### Resultado:

* La nueva pendiente es $a' = \frac{a}{h}$
* El nuevo término independiente $b'$ también **cambia**, pero no de forma tan simple (depende de $a$, $b$, y los promedios).

#### Interpretación práctica:

> Si **escalás todos los valores de entrada $x_i$** por un factor $h$, entonces **la pendiente se divide por $h$**.

#### Ejemplo concreto:

* Si antes tu modelo era:

  $$
  y = 4x + 2
  $$
* Y escalás $x \rightarrow 2x$, la pendiente se convierte en:

  $$
  y = \frac{4}{2}x + b'
  \Rightarrow y = 2x + b'
  $$

(El término $b'$ se ajusta también, pero su fórmula es más compleja).

---

#### ¿Para qué sirve esto?

* Esto explica por qué **la escala de las variables afecta la regresión**.
* Si los $x_i$ están en segundos, minutos o años, **la pendiente será diferente**, aunque el comportamiento siga siendo el mismo.
* Por eso es común **normalizar o estandarizar los datos** antes de aplicar regresión.


7 – Interpretar (y/o graficar en forma aproximada) las siguientes seis situaciones para
regresión lineal: a < 0, a > 0, a = 0, a < 1, a > 1, a = 1 


$$
y = ax + b
$$

Nos centramos en cómo varía la pendiente $a$ (y no el término independiente $b$, que solo desplaza la recta verticalmente).


#### 🔻 CASO 1: $a < 0$

📉 **Recta descendente**

* A medida que $x$ aumenta, $y$ disminuye.
* Ejemplo: $y = -2x + 3$

**Interpretación práctica**: relación negativa entre variables. Si una sube, la otra baja.


#### 🔺 CASO 2: $a > 0$

📈 **Recta ascendente**

* A medida que $x$ aumenta, $y$ también aumenta.
* Ejemplo: $y = 2x + 1$

**Interpretación práctica**: relación positiva entre variables. Si una sube, la otra también.

#### ➖ CASO 3: $a = 0$

📏 **Recta horizontal**

* El valor de $y$ no cambia aunque $x$ varíe.
* Ejemplo: $y = 5$

**Interpretación práctica**: no hay relación entre las variables.

#### 🔽 CASO 4: $0 < a < 1$

📈 **Recta ascendente suave**

* Crece lentamente.
* Ejemplo: $y = 0.5x + 2$

**Interpretación práctica**: relación positiva débil; por cada unidad que crece $x$, $y$ crece menos de 1.

#### 🔼 CASO 5: $a > 1$

📈 **Recta ascendente empinada**

* Crece rápidamente.
* Ejemplo: $y = 3x + 1$

**Interpretación práctica**: relación positiva fuerte; pequeños cambios en $x$ producen grandes cambios en $y$.


#### 🔸 CASO 6: $a = 1$

📈 **Crecimiento perfecto uno a uno**

* Ejemplo: $y = x$

**Interpretación práctica**: por cada unidad que crece $x$, $y$ también crece exactamente una unidad.
