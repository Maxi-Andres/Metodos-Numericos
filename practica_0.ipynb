1. Implementar en Python el proceso anterior, que use una función (externa implementada, o bien pasada como parámetro), que lea el x como parámetro y un n natural, y luego sobre el ex inicial aplique la f n veces, finalmente devolviendo el resultado obtenido tras la última aplicación.



```python
import math
import matplotlib.pyplot as plt

# Definición de funciones
func1 = lambda x: x**2 - 1
func2 = lambda x: (x + 3) / 2
func3 = lambda x: math.exp(-x)
func4 = math.cos

# Valor inicial
x0 = 1.0
xa = xb = xc = xd = x0  # Asignamos el mismo valor inicial a todas

n_iteraciones = 15

# Listas para guardar los valores de cada función
valores1 = [xa]
valores2 = [xb]
valores3 = [xc]
valores4 = [xd]

# Iteración
for i in range(n_iteraciones):
    xa = func1(xa)
    xb = func2(xb)
    xc = func3(xc)
    xd = func4(xd)

    valores1.append(xa)
    valores2.append(xb)
    valores3.append(xc)
    valores4.append(xd)

# Mostrar resultados por consola
for i in range(len(valores1)):
    print(f"Iteración {i}: f1={valores1[i]:.6f}, f2={valores2[i]:.6f}, f3={valores3[i]:.6f}, f4={valores4[i]:.6f}")

# Graficar los resultados
plt.plot(valores1, label="x^2 - 1", marker='o')
plt.plot(valores2, label="(x + 3)/2", marker='s')
plt.plot(valores3, label="exp(-x)", marker='^')
plt.plot(valores4, label="cos(x)", marker='x')

plt.title("Iteración de distintas funciones")
plt.xlabel("Iteración")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt
import itertools

# Lista de funciones con nombres
func_list = [
    ("x^2 - 1", lambda x: x**2 - 1),
    ("(x + 3)/2", lambda x: (x + 3) / 2),
    ("cos(x)", lambda x: math.cos(x)),
    ("exp(-x)", lambda x: math.exp(-x)),
    ("1 / (1 + x)", lambda x: 1 / (1 + x) if x != -1 else float('inf')),
    ("sqrt(abs(x))", lambda x: math.sqrt(abs(x))),
    ("tanh(x)", lambda x: math.tanh(x)),
    ("abs(x - 2)", lambda x: abs(x - 2)),
    ("sin(x)", lambda x: math.sin(x))
]

# Valores iniciales
x0 = 1.0
n_iteraciones = 20

# Todas las combinaciones de funciones de a dos
combinaciones = list(itertools.combinations(func_list, 2))

# Parámetros para las grillas
figs_por_fila = 6
figs_por_figura = figs_por_fila ** 2
total = len(combinaciones)

# Colores y estilos de línea variados
colores = ['red', 'blue', 'green', 'orange', 'purple', 'cyan', 'black', 'magenta', 'brown']
estilos = ['-', '--', '-.', ':']

for i in range(0, total, figs_por_figura):
    fig, axes = plt.subplots(figs_por_fila, figs_por_fila, figsize=(15, 10))
    axes = axes.flatten()

    for j in range(figs_por_figura):
        if i + j >= total:
            break
        (name1, f1), (name2, f2) = combinaciones[i + j]
        x = x0
        valores = [x]

        for k in range(n_iteraciones):
            try:
                x = f1(x) if k % 2 == 0 else f2(x)
                if abs(x) > 1e6:
                    valores.append(float('inf'))
                    break
                valores.append(x)
            except:
                valores.append(float('nan'))
                break

        # Estado de comportamiento
        if any(math.isnan(v) or v == float('inf') for v in valores):
            estado = "Diverge/Error"
        elif max(valores) - min(valores) < 0.0001:
            estado = "Converge"
        elif len(set(round(v, 5) for v in valores[-5:])) < 3:
            estado = "Cicla"
        else:
            estado = "Mixto"

        # Selección de color y estilo
        color = colores[j % len(colores)]
        estilo = estilos[(j // len(colores)) % len(estilos)]

        # Graficar
        ax = axes[j]
        ax.plot(valores, linestyle=estilo, color=color, marker='o', markersize=3)
        ax.set_title(f"{name1} + {name2}\n{estado}", fontsize=8)
        ax.tick_params(labelsize=6)
        ax.grid(True)

    # Ocultar subplots vacíos
    for k in range(j + 1, figs_por_figura):
        fig.delaxes(axes[k])

    fig.tight_layout()
    plt.show()
```

4. Aplicar el método de bisección para hallar una raíz aproximada de la función f(x) = x^3 + x - 5, comenzando con el intervalo [1,2], y el error epsilon = 0.5.



```python
def f(x):
    return x**3 + x - 5

def biseccion(a, b, epsilon):
    if f(a) * f(b) >= 0:
        print("Error: f(a) y f(b) deben tener signos opuestos.") # si lo evaluas y ambos son positivos o negativos es que no hay raiz entre ellos, ESTE METODO SOLO SIRVE SI LA IMAGEN TIENE NEGATIVOS Y POSITIVOS
        return None

    while ((b - a) >= epsilon):
        c = (a + b) / 2
        if f(c) == 0 or (b - a) < epsilon:
            return c
        elif f(c) * f(a) < 0: # cuidado con esto porque se puede hacer muy grande
            b = c
        else:
            a = c
    return (a + b) / 2 # devuelvo el punto medio del rango si es mas chico que el epsilon y no encontre la raiz

# Entrada de datos
a = float(input("Ingrese el primer valor (a): "))
b = float(input("Ingrese el segundo valor (b): "))
epsilon = float(input("Ingrese el valor de epsilon: "))

# Cálculo y resultado
raiz = biseccion(a, b, epsilon)
if raiz is not None:
    print(f"La raíz aproximada es: {raiz}")
```

5. Aplicar el método de bisección para hallar una raíz aproximada de la función f(x) = x^3 + x - 10 haciendo 4 pasos.



```python
def f(x):
    return x**3 + x - 10

def biseccion(a, b, epsilon, maxStep):
    if f(a) * f(b) >= 0:
        print("Error: f(a) y f(b) deben tener signos opuestos.") # si lo evaluas y ambos son positivos o negativos es que no hay raiz entre ellos
        return None

    step = 0
    print("\nIteración\t a\t\t b\t\t c\t\t f(c)")
    print("------------------------------------------------------------")
    while (b - a >= epsilon) and (step <= maxStep):
        c = (a + b) / 2
        print(f"{step}\t\t {a:.6f}\t {b:.6f}\t {c:.6f}\t {f(c):.6f}")
        
        if f(c) == 0 or (b - a) < epsilon:
            print(f"\nSe encontró una raíz (o se alcanzó la precisión) en el paso {step}.")
            return c
        elif f(c) * f(a) < 0: # cuidado con esto porque se puede hacer muy grande
            b = c
        else:
            a = c

        step += 1

    print(f"\nSe alcanzó el máximo número de pasos ({maxStep}) o precisión requerida.")
    return (a + b) / 2 # devuelvo el punto medio del rango si es mas chico que el epsilon y no encontre la raiz

# Entrada de datos
a = float(input("Ingrese el primer valor (a): "))
b = float(input("Ingrese el segundo valor (b): "))
epsilon = float(input("Ingrese el valor de epsilon: "))
maxStep = 4  # Podes cambiar este valor si querés más pasos

# Cálculo y resultado
raiz = biseccion(a, b, epsilon, maxStep)
if raiz is not None:
    print(f"\nLa raíz aproximada es: {raiz}")
```

6. Aplicar el método de bisección para hallar una aproximación de un x que cumpla x^2 + x = 12. Calcular luego el valor exacto de otra manera, y comparar.



```python
def f(x):
    return x**2 + x - 12

def biseccion(a, b, epsilon, maxStep):
    if f(a) * f(b) >= 0:
        print("Error: f(a) y f(b) deben tener signos opuestos.") # si lo evaluas y ambos son positivos o negativos es que no hay raiz entre ellos
        return None

    step = 0
    print("\nIteración\t a\t\t b\t\t c\t\t f(c)")
    print("------------------------------------------------------------")
    while (b - a >= epsilon) and (step <= maxStep):
        c = (a + b) / 2
        print(f"{step}\t\t {a:.6f}\t {b:.6f}\t {c:.6f}\t {f(c):.6f}")
        
        if f(c) == 0 or (b - a) < epsilon:
            return c
        elif f(c) * f(a) < 0: # cuidado con esto porque se puede hacer muy grande
            b = c
        else:
            a = c
        step += 1

    return (a + b) / 2 # devuelvo el punto medio del rango si es mas chico que el epsilon y no encontre la raiz

# Parametros
a = 1
b = 6
epsilon = 0.0001
maxStep = 100

# Bisección
raiz_aprox = biseccion(a, b, epsilon, maxStep)
print(f"\nRaíz aproximada por bisección: {raiz_aprox:.6f}")

# Solucion con formula cuadratica
import math
x1 = (-1 + math.sqrt(1**2 + 4*12)) / 2  # (-b + √(b² - 4ac)) / 2a
x2 = (-1 - math.sqrt(1**2 + 4*12)) / 2

print(f"Raíces exactas: x1 = {x1:.6f}, x2 = {x2:.6f}")

# Comparación
error = abs(x1 - raiz_aprox)
print(f"Error absoluto entre la raíz exacta (x1) y la aproximada: {error:.6f}")
```

Implementar en Python un método para aproximar raíces al estilo de bisección, pero que en vez de dividir el intervalo en 2 subintervalos lo divida en 3, y en cada iteración elija uno de los 3 subintervalos que contenga una raíz.
```python
def f(x):
    return x**2 + x - 12  # Ejemplo: resolver x^2 + x = 12

def triseccion(a, b, epsilon, maxStep):
    Fa = f(a)
    Fb = f(b)

    if Fa * Fb >= 0:
        print("Error: f(a) y f(b) deben tener signos opuestos.")
        return None

    step = 0
    print("\nIteración\t a\t\t c1\t\t c2\t\t b\t\t f(c1)\t\t f(c2)")
    print("-----------------------------------------------------------------------------------------------------------")

    while (b - a >= epsilon) and (step <= maxStep):
        c1 = a + (b - a) / 3
        c2 = b - (b - a) / 3
        Fc1 = f(c1)
        Fc2 = f(c2)

        print(f"{step}\t\t {a:.6f}\t {c1:.6f}\t {c2:.6f}\t {b:.6f}\t {Fc1:.6f}\t {Fc2:.6f}")

        if Fc1 == 0 or (b - a) < epsilon:
            print(f"\nSe encontró una raíz (o se alcanzó la precisión) en el paso {step}.")
            return c1
        if Fc2 == 0 or (b - a) < epsilon:
            print(f"\nSe encontró una raíz (o se alcanzó la precisión) en el paso {step}.")
            return c2
        
        if Fc1 * Fa < 0: # cuidado con esto porque se puede hacer muy grande
            b = c1
            Fb = Fc1 # de mas creo
        elif Fc1 * Fc2 < 0:
            a = c1
            b = c2 
            Fa = Fc1 # de mas creo
            Fb = Fc2 # de mas creo
        else:
            a = c2
            Fa = Fc2 # de mas creo

        step += 1

    if(step == maxStep):
        print(f"\nSe alcanzó el máximo número de pasos ({maxStep})")
    else:
        print(f"\nSe alcanzó la precisión requerida.")
    return (a + b) / 2 # devuelvo el punto medio del rango si es mas chico que el epsilon y no encontre la raiz

# Parámetros
a = 0
b = 7
epsilon = 0.0001
maxStep = 100

# Cálculo y resultado
raiz = triseccion(a, b, epsilon, maxStep)
if raiz is not None:
    print(f"\nLa raíz aproximada es: {raiz}")
```
